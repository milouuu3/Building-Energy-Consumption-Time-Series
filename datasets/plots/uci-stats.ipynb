```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(
    "../uci/household_power_consumption.txt",
    sep=";",
    parse_dates={"timestamp": ["Date", "Time"]},
    na_values="?",
    infer_datetime_format=True,
    low_memory=False,
    index_col="timestamp"
)
```

```python
# Plot over time

plt.figure(figsize=(12, 6))
plt.plot(data.index, data["Global_active_power"], color="blue", linewidth=0.5)
plt.title("Global_active_power over time")
plt.xlabel("Time")
plt.ylabel("Global_active_power (kW)")
plt.tight_layout()
plt.grid()
plt.savefig("uci/uci_overtime.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
# Plot missing values of target

missing = data["Global_active_power"].isna().astype(int)
plt.figure(figsize=(12, 4))
missing.plot(color="red", alpha=0.5, linewidth=0.5)
plt.title("Missing Values in target variable Global_active_power")
plt.xlabel("Timestamp")
plt.ylabel("(1 = missing, 0 = present)")
plt.tight_layout()
plt.savefig("uci/uci_missing.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
