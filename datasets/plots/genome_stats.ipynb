```python
import pandas as pd
import matplotlib.pyplot as plt

metadata = pd.read_csv("../genome/metadata.csv")
electricity = pd.read_csv("../genome/electricity_cleaned.csv", parse_dates=True, index_col='timestamp')
```

```python
# Number of Buildings per Primary Use Type

metadata['primaryspaceusage'].value_counts().plot(kind='barh', figsize=(8, 6), color="blue")
plt.title("Number of Buildings per primary use category")
plt.xlabel("Number of Buildings")
plt.tight_layout()
plt.grid()
plt.savefig("genome/primary_use.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
# Calculate missing value fraction per building

missing_fraction = electricity.isna().mean()

plt.figure(figsize=(8, 4))
missing_fraction.hist(bins=20, color="blue", edgecolor="black")
plt.title("Distribution of Missing Value Fraction for all buildings")
plt.xlabel("Fraction of Missing Values")
plt.ylabel("Number of Buildings")
plt.tight_layout()
plt.savefig("genome/missing_fractions.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
# Example building

name = 'Panther_office_Hannah'
office_example = pd.DataFrame(electricity[name].truncate(before='2017-01-01'))
office_example.plot(figsize=(12,6), color="blue")
plt.title(f"{name} over time")
plt.xlabel("Timestamp")
plt.ylabel("Energy Consumption (kWh)")
plt.tight_layout()
plt.grid()
plt.savefig("genome/example_hannah.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
