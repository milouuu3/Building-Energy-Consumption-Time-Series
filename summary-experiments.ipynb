```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
maxwell_imputation = pd.read_csv("performance/summary/maxwell-imputation.csv")
maxwell_forecasting = pd.read_csv("performance/summary/maxwell-forecasting.csv")
maxwell_codecarbon = pd.read_csv("performance/summary/maxwell-codecarbon.csv")
maxwell_eco2ai = pd.read_csv("performance/summary/maxwell-eco2ai.csv")
```

```python
maxwell_imputation_avg = maxwell_imputation.groupby("Method", as_index=False)[["MAE", "NRMSE"]].mean()
maxwell_forecasting_avg = maxwell_forecasting.groupby("Method", as_index=False)[["MAE"]].mean()
maxwell_codecarbon_avg = maxwell_codecarbon.groupby("Method", as_index=False)[["Runtime", "CO2 emissions", "Energy Consumption"]].mean()
maxwell_eco2ai_avg = maxwell_eco2ai.groupby("Method", as_index=False)[["Runtime", "CO2 emissions", "Energy Consumption"]].mean()
```

```python
df = maxwell_imputation_avg.set_index("Method")

df.plot(kind="bar", figsize=(10, 6), rot=45)
plt.ylabel("MAE and NRMSE on Grid Import (log scale) ")
plt.yscale("log")
plt.title("Imputation Accuracy Average MAE and NRMSE by Method")
plt.tight_layout()
plt.savefig("maxwell-imputation.png")
```

```python
df = maxwell_forecasting_avg.set_index("Method")

df.plot(kind="bar", figsize=(10, 6), rot=45)
plt.ylabel("MAE on Grid Import (log scale) ")
plt.yscale("log")
plt.title("Forecasting Average MAE by Method")
plt.tight_layout()
plt.savefig("maxwell-forecasting.png")
```

```python
df = maxwell_codecarbon_avg.set_index("Method")

df.plot(kind="bar", figsize=(10, 6), rot=45)
plt.ylabel("(log scale)")
plt.yscale("log")
plt.title("CodeCarbon Average Performance by Method")
plt.legend(["Runtime (s)", "CO2 Emissions (kg)", "Energy Consumption (kWh)"])
plt.tight_layout()
plt.savefig("maxwell-codecarbon.png")
```

```python
df = maxwell_eco2ai_avg.set_index("Method")

df.plot(kind="bar", figsize=(10, 6), rot=45)
plt.ylabel("(log scale)")
plt.yscale("log")
plt.title("ecoaAI Average Performance by Method")
plt.legend(["Runtime (s)", "CO2 Emissions (kg)", "Energy Consumption (kWh)"])
plt.tight_layout()
plt.savefig("maxwell-eco2ai.png")
```
